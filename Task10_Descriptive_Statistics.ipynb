## **QQ- Plots**

A q-q plot is a plot of the quantiles of the first data set against the quantiles of the second data set.

**Mthly_HH_Income**

```python
fig, ax = plt.subplots(figsize=(10,6))
Mthly_HH_Income = data.Mthly_HH_Income
sns.distplot(Mthly_HH_Income)
```

```python
from scipy.stats import probplot
import pylab
```

```python
fig, ax = plt.subplots(figsize=(10,6))
A = probplot(Mthly_HH_Income,dist='norm',plot=pylab)
pylab.show(A)
```

**Mthly_HH_Expense**

```python
Mthly_HH_Expense = data.Mthly_HH_Expense
fig, ax = plt.subplots(figsize=(10,6))
sns.distplot(Mthly_HH_Expense)
B = probplot(Mthly_HH_Expense,dist='norm',plot=pylab)
pylab.show(B)
```
